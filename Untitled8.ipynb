#*Аналіз даних Penguins за допомогою бібліотеки Seaborn*

Мета роботи: ознайомитися з бібліотекою Seaborn, побудувати графіки для аналізу набору даних **penguins**, дослідити взаємозв'язок між ознаками та зробити висновки.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Завантаження набору даних
penguins = sns.load_dataset("penguins")

# Виведення перших 5 рядків
print("Перші 5 рядків набору даних:")
display(penguins.head())

print("Розмір датасету:", penguins.shape)

print("\nІнформація про дані:")
penguins.info()
```

## 1. Аналіз розподілу ваги та висоти птахів

```python
# Видаляємо рядки з пропущеними значеннями
data = penguins.dropna()

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=data,
    x="bill_length_mm",
    y="body_mass_g",
    hue="species",
    style="species",
    s=80
)

plt.title("Залежність маси тіла від довжини дзьоба")
plt.xlabel("Довжина дзьоба (мм)")
plt.ylabel("Маса тіла (г)")
plt.grid(True)

plt.show()
```

### **Висновок**

На графіку видно позитивну залежність між довжиною дзьоба та масою тіла пінгвінів. Різні види утворюють окремі групи, що свідчить про відмінності між ними.

## 2. Вивчення впливу виду птаха на розміри крил

```python
plt.figure(figsize=(8,6))

sns.boxplot(
    data=data,
    x="species",
    y="flipper_length_mm",
    hue="species",
    legend=False
)

plt.title("Довжина крил залежно від виду пінгвіна")
plt.xlabel("Вид")
plt.ylabel("Довжина крила (мм)")

plt.show()
```

###**Висновок**

Пінгвіни виду **Gentoo** мають найбільшу довжину крил, **Adelie** — найменшу, а **Chinstrap** займають проміжне положення.

## 3. Кореляція між різними ознаками

```python
numeric = data.select_dtypes(include="number")

corr = numeric.corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Теплова карта кореляції")

plt.show()
```

```python
print("Матриця кореляції:")
display(corr)
```

### Аналіз кореляції

Найсильніша позитивна кореляція спостерігається між:

- **body_mass_g** та **flipper_length_mm**;
- **body_mass_g** та **bill_length_mm**;
- **bill_length_mm** та **flipper_length_mm**.

Це означає, що більші за масою пінгвіни, як правило, мають довші крила та довший дзьоб.
