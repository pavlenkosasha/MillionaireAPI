```python
# Импорт библиотеки
import pandas as pd

# Загрузка данных
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

column_names = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
    "class"
]

iris_df = pd.read_csv(url, header=None, names=column_names)

# Просмотр первых строк
print("Первые 5 строк:")
display(iris_df.head())
```

<div align="center">

**1️⃣ Аналіз даних Iris**

*⭐1.1 Середня довжина чашолистика (sepal length) для кожного виду⭐*


```python
print("Середня довжина чашолистика:")
display(iris_df.groupby("class")["sepal_length"].mean())
```

*⭐1.2 Максимальна ширина пелюстки (petal width) для виду setosa⭐*

```python
max_width_setosa = iris_df[iris_df["class"] == "Iris-setosa"]["petal_width"].max()

print("Максимальна ширина пелюстки для setosa:")
print(max_width_setosa)
```

*⭐1.3 Розподіл довжини пелюстки (petal length)⭐*

```python
print("Опис розподілу довжини пелюстки:")
display(iris_df["petal_length"].describe())
```

<div align="center">

**2️⃣Фільтрація та відбір даних**

*✅2.1 Дані лише для виду versicolor✅*

```python
versicolor_df = iris_df[iris_df["class"] == "Iris-versicolor"]

print("Versicolor:")
display(versicolor_df.head())
```

*✅2.2 Іриси з довжиною пелюстки більше 5.0✅*

```python
long_petal = iris_df[iris_df["petal_length"] > 5.0]

print("Petal length > 5.0")
display(long_petal.head())
```

<div align="center">

**3️⃣Групування та агрегація**

*✨3.1 Середня ширина пелюстки для кожного виду✨*

```python
print("Середня ширина пелюстки:")
display(iris_df.groupby("class")["petal_width"].mean())
```

*✨3.2 Мінімальна довжина чашолистика для кожного виду✨*

```python
print("Мінімальна довжина чашолистика:")
display(iris_df.groupby("class")["sepal_length"].min())
```

*✨3.3 Кількість ірисів кожного виду, у яких довжина пелюстки більша за середню✨*

```python
average_petal = iris_df["petal_length"].mean()

result = (
    iris_df[iris_df["petal_length"] > average_petal]
    .groupby("class")
    .size()
)

print("Кількість ірисів з petal_length > середнього:")
display(result)
```
