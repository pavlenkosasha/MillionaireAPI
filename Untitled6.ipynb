# **Завдання 1. Створення масивів та операції**

```python
import numpy as np


# Щоб результати були однаковими при кожному запуску
np.random.seed(42)

# Створення двох масивів 5x5
A = np.random.randint(1, 10, (5, 5))
B = np.random.randint(1, 5, (5, 5))  # невеликі числа для степенів

print("Масив A:")
print(A)

print("\nМасив B:")
print(B)

# Сума елементів
print("\nСума елементів A:", np.sum(A))
print("Сума елементів B:", np.sum(B))

# Різниця
print("\nРізниця A - B:")
print(A - B)

# Поелементне множення
print("\nДобуток A * B:")
print(A * B)

# Піднесення до степеня
print("\nA ** B:")
print(np.power(A, B))
```

# Завдання 2. Індексація та вибірка даних

```python
import numpy as np

# Створення шахової дошки 8x8
board = np.zeros((8, 8), dtype=int)
board[1::2, ::2] = 1
board[::2, 1::2] = 1

print("Шахова дошка:")
print(board)

# 3-й рядок
print("\n3-й рядок:")
print(board[2])

# 5-й стовпець
print("\n5-й стовпець:")
print(board[:, 4])

# Лівий верхній кут 3x3
print("\nПідмасив 3x3:")
print(board[:3, :3])
```

# Завдання 3. Робота із зображенням та статистика

```python
from google.colab import files
from PIL import Image
import numpy as np

# Завантаження файлу
uploaded = files.upload()
```

```python
# Отримання назви файлу
filename = list(uploaded.keys())[0]

# Відкриття зображення
img = Image.open(filename).convert("RGB")

# Перетворення в NumPy-масив
img_array = np.array(img)

print("Розмір масиву:", img_array.shape)
print("Тип даних:", img_array.dtype)

# Статистика по каналах
channels = ["R", "G", "B"]

for i, ch in enumerate(channels):
    print(f"\nКанал {ch}")
    print("Середнє:", np.mean(img_array[:, :, i]))
    print("Мінімум:", np.min(img_array[:, :, i]))
    print("Максимум:", np.max(img_array[:, :, i]))

# Загальна сума інтенсивностей
print("\nЗагальна сума інтенсивності:")
print(np.sum(img_array))

# Нормалізація
normalized = img_array / np.max(img_array)

print("\nНормалізоване зображення")
print("Мінімум:", normalized.min())
print("Максимум:", normalized.max())
```
