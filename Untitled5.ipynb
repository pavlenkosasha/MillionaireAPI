```python
import numpy as np

# Створюємо два масиви 5x5 з випадковими числами від 1 до 9
array1 = np.random.randint(1, 10, (5, 5))
array2 = np.random.randint(1, 10, (5, 5))

# Виводимо масиви
print("Перший масив:")
print(array1)

print("\nДругий масив:")
print(array2)

# Обчислюємо суму елементів
sum1 = np.sum(array1)
sum2 = np.sum(array2)

print("\nСума елементів першого масиву:", sum1)
print("Сума елементів другого масиву:", sum2)
```

```python
# Різниця між елементами двох масивів
difference = array1 - array2

print("Різниця між масивами:")
print(difference)
```

```python
a = [2, 3, 4]
b = [5, 6, 7]

# Перемноження
multiply = [x * y for x, y in zip(a, b)]
print(multiply)

# Піднесення до степеня
power = [x ** y for x, y in zip(a, b)]
print(power)
```
